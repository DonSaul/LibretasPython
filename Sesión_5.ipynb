# Introducción al Módulo Math en Python

El módulo `math` es una biblioteca estándar en Python que proporciona funciones y constantes matemáticas para realizar operaciones numéricas avanzadas. Con `math`, puedes realizar cálculos trigonométricos, logarítmicos, exponenciales y mucho más. Veamos algunas de las funcionalidades principales que ofrece este módulo:

## Funciones Trigonométricas:
- `math.sin(x)`: Calcula el seno de x en radianes.
- `math.cos(x)`: Calcula el coseno de x en radianes.
- `math.tan(x)`: Calcula la tangente de x en radianes.

## Funciones Logarítmicas:
- `math.log(x)`: Calcula el logaritmo natural (base e) de x.
- `math.log2(x)`: Calcula el logaritmo en base 2 de x.

## Otras Funciones Útiles:
- `math.sqrt(x)`: Calcula la raíz cuadrada de x.
- `math.exp(x)`: Calcula e elevado a la potencia x.
- `math.pow(x, y)`: Calcula x elevado a la potencia y.

## Constantes Importantes:
- `math.pi`: El valor de π (pi).
- `math.e`: El valor de e (la base del logaritmo natural).

## Ejemplos de Uso:



```python
import math

# Calcular el seno de 45 grados
seno_45 = math.sin(math.radians(45))
print("El seno de 45 grados es:", seno_45)

# Calcular el coseno de π/4 radianes
coseno_pi_4 = math.cos(math.pi / 4)
print("El coseno de π/4 radianes es:", coseno_pi_4)
```

# Introducción a NumPy en Python

NumPy es una biblioteca fundamental en Python para computación numérica. Proporciona una estructura de datos de matriz multidimensional eficiente, junto con una amplia gama de funciones para operar en estas matrices. Veamos algunas de las funcionalidades principales que ofrece NumPy:

## Creación de Arrays:
- `numpy.array()`: Crea un array NumPy a partir de una lista o tupla.
- `numpy.zeros()`: Crea un array NumPy lleno de ceros.
- `numpy.ones()`: Crea un array NumPy lleno de unos.
- `numpy.arange()`: Crea un array NumPy con valores espaciados uniformemente dentro de un intervalo.

## Operaciones Básicas de Arrays:
- Suma, resta, multiplicación y división de arrays.
- Operaciones matriciales como la transposición, el producto punto y la inversa.

## Funciones Universales (ufuncs):
- Funciones matemáticas que operan elemento por elemento en un array, como `numpy.sin()`, `numpy.cos()`, `numpy.exp()`, etc.

## Indexación y Segmentación:
- Acceder a elementos individuales o subconjuntos de un array.
- Utilización de indexación booleana para seleccionar elementos basados en condiciones.

## Ejemplos de Uso:

### Creación de Arrays:




```python
import numpy as np

# Crear un array de una dimensión
arr1d = np.array([1, 2, 3, 4, 5])
print("Array 1D:", arr1d)

# Crear un array de dos dimensiones lleno de ceros
arr2d_zeros = np.zeros((2, 3))
print("Array 2D de ceros:\n", arr2d_zeros)

# Crear un array de tres dimensiones lleno de unos
arr3d_ones = np.ones((3, 2, 2))
print("Array 3D de unos:\n", arr3d_ones)

# Crear un array de valores espaciados uniformemente
arr_range = np.arange(0, 10, 2)
print("Array con valores espaciados uniformemente:", arr_range)
```

##1. Suma, Resta, Multiplicación y División de Arrays:




* Puedes realizar operaciones aritméticas elemento por elemento entre dos arrays NumPy del mismo tamaño.
* Estas operaciones se realizan de forma vectorizada, lo que significa que no necesitas bucles para aplicar las operaciones a cada elemento individualmente.

```python
import numpy as np

# Definir dos arrays NumPy
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Suma de arrays
suma = arr1 + arr2
print("Suma de arrays:", suma)

# Resta de arrays
resta = arr1 - arr2
print("Resta de arrays:", resta)

# Multiplicación de arrays
multiplicacion = arr1 * arr2
print("Multiplicación de arrays:", multiplicacion)

# División de arrays
division = arr2 / arr1
print("División de arrays:", division)
```

## Matrices en Python

Una matriz en Python es una estructura de datos bidimensional que organiza los datos en filas y columnas, similar a una tabla. Cada elemento en la matriz tiene una posición única definida por su fila y columna.

### Ejemplo:

Supongamos que tenemos una matriz de 2x3 que representa las calificaciones de los estudiantes en tres asignaturas:

| Matemáticas | Ciencias | Historia |
|-------------|----------|----------|
| 85          | 90       | 75       |
| 70          | 80       | 85       |

Cada número en la tabla representa la calificación de un estudiante en una asignatura específica. Por ejemplo, el número 90 representa la calificación del segundo estudiante en Ciencias.

En Python, podemos representar esta matriz utilizando NumPy de la siguiente manera:




```python
import numpy as np

# Definir la matriz de calificaciones
calificaciones = np.array([[85, 90, 75], [70, 80, 85]])

# Imprimir la matriz
print("Matriz de calificaciones:")
print(calificaciones)
```


##2. Operaciones Matriciales:
* NumPy proporciona funciones para realizar operaciones matriciales como la
transposición, el producto punto, la inversa, etc.
* Estas operaciones son útiles en álgebra lineal y en muchas aplicaciones científicas y de ingeniería.


```python
import numpy as np

# Definir dos matrices NumPy
matriz1 = np.array([[1, 2], [3, 4]])
matriz2 = np.array([[5, 6], [7, 8]])

# Producto punto de matrices
producto_punto = np.dot(matriz1, matriz2)
print("Producto punto de matrices:", producto_punto)

# Transposición de una matriz
transpuesta = matriz1.T
print("Transpuesta de la matriz:", transpuesta)
```

##3. Operaciones con Escalares:
Puedes realizar operaciones aritméticas entre un escalar y un array NumPy. Estas operaciones se aplicarán a cada elemento del array.

```python
import numpy as np

# Definir un array NumPy
arr = np.array([1, 2, 3])

# Suma de un escalar al array
suma_escalar = arr + 1
print("Suma de un escalar al array:", suma_escalar)

# Multiplicación de un escalar al array
multiplicacion_escalar = arr * 2
print("Multiplicación de un escalar al array:", multiplicacion_escalar)
```

# Matplotlib en Python

Matplotlib es una biblioteca en Python ampliamente utilizada para la visualización de datos. Proporciona una amplia variedad de funciones para crear gráficos de alta calidad, como gráficos de líneas, gráficos de dispersión, histogramas, gráficos de barras, entre otros. Matplotlib es flexible y altamente personalizable, lo que la hace adecuada para una amplia gama de aplicaciones de visualización.

### Características Principales:

1. **Versatilidad:** Matplotlib es capaz de crear una amplia variedad de tipos de gráficos, desde gráficos simples hasta gráficos complejos.

2. **Personalización:** Permite personalizar todos los aspectos de los gráficos, incluyendo etiquetas, títulos, colores, estilos de línea, marcadores, etc.

3. **Interactividad:** Matplotlib es compatible con interactividad en los gráficos cuando se utiliza en Jupyter Notebooks o en aplicaciones interactivas.




```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear un gráfico de líneas
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Datos')

# Añadir etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

##Ejemplo
###1. Importar Matplotlib y Datos:
Primero, importamos la biblioteca Matplotlib y cualquier otra biblioteca necesaria, como NumPy para generar datos, y luego definimos los datos que queremos visualizar.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = np.linspace(0, 10, 100)  # Creamos 100 puntos entre 0 y 10
y = np.sin(x)  # Calculamos el seno de cada punto
```

###2. Crear el Gráfico:
Usamos funciones de Matplotlib, como plt.plot(), para crear el gráfico en función de nuestros datos. En este ejemplo, creamos un gráfico de línea utilizando los datos x e y.

```python
# Crear el gráfico de línea
plt.plot(x, y, label='sin(x)')
```

###3. Personalizar el Gráfico:
Podemos personalizar el aspecto del gráfico ajustando los colores, el estilo de línea, los marcadores, los límites de los ejes, las etiquetas, el título, etc.

```python
# Personalizar el gráfico

plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('sin(x)')  # Etiqueta del eje y
plt.title('Gráfico de la Función Seno')  # Título del gráfico
plt.legend()  # Mostrar leyenda
```

###4. Mostrar el Gráfico:
Finalmente, utilizamos `plt.show()` para mostrar el gráfico en una ventana emergente o en la salida del notebook.

```python
# Mostrar el gráfico
plt.show()
```

###5. Código Completo

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = np.linspace(0, 10, 100)  # Creamos 100 puntos entre 0 y 10
y = np.sin(x)  # Calculamos el seno de cada punto

# Crear el gráfico de línea
plt.plot(x, y, label='sin(x)')

# Personalizar el gráfico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('sin(x)')  # Etiqueta del eje y
plt.title('Gráfico de la Función Seno')  # Título del gráfico
plt.legend()  # Mostrar leyenda

# Mostrar el gráfico
plt.show()
```

##Gráficos Variados
###1. Gráfico de Barras Horizontales:

```python
import matplotlib.pyplot as plt

# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [10, 20, 15, 25, 30]

# Crear el gráfico de barras horizontales
plt.barh(categorias, valores, color='skyblue')

# Personalizar el gráfico
plt.xlabel('Valores')
plt.ylabel('Categorías')
plt.title('Gráfico de Barras Horizontales')

# Mostrar el gráfico
plt.show()
```

###2. Gráfico de Dispersión:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = np.random.randn(100)
y = np.random.randn(100)

# Crear el gráfico de dispersión
plt.scatter(x, y, color='orange')

# Personalizar el gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Dispersión')

# Mostrar el gráfico
plt.show()
```

###3. Gráfico de Pastel:

```python
import matplotlib.pyplot as plt

# Datos
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

# Crear el gráfico de pastel
plt.pie(sizes, labels=labels, colors=['red', 'green', 'blue', 'purple', 'orange'])

# Personalizar el gráfico
plt.title('Gráfico de Pastel')

# Mostrar el gráfico
plt.show()
```

###4. Gráfico de Barras Apiladas:


```python
import matplotlib.pyplot as plt

# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [10, 15, 20, 25, 30]
valores2 = [5, 10, 15, 20, 25]

# Crear el gráfico de barras apiladas
plt.bar(categorias, valores1, color='purple', label='Grupo 1')
plt.bar(categorias, valores2, bottom=valores1, color='orange', label='Grupo 2')

# Personalizar el gráfico
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras Apiladas')
plt.legend()

# Mostrar el gráfico
plt.show()
```
